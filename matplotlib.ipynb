```python
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
%matplotlib inline
```

```python
stock_prices_apple = [93.95, 112.14, 104.05, 144.85, 169.49]
stock_prices_microsoft = [39.01, 50.29, 57.05, 69.98, 94.39]
years = [2014, 2015, 2016, 2017, 2018]
```

```python
plt.rcParams['figure.facecolor'] = '#182630'
plt.rcParams['axes.facecolor'] = '#4c4a48'
plt.plot(years, stock_prices_apple, 'k', years, stock_prices_microsoft, 'b')

plt.xlabel('Year')
plt.ylabel('Stock price')

plt.axis([2013, 2019, 10, 180])

plt.show()
```

```python
fig = plt.figure(1, figsize=(6.4, 4.8))
apple_stocks = fig.add_subplot(121)
microsoft_stocks = fig.add_subplot(122)

apple_stocks.plot(years, stock_prices_apple)
apple_stocks.xaxis.set_major_locator(MaxNLocator(integer=True))

microsoft_stocks.scatter(years, stock_prices_microsoft)
microsoft_stocks.xaxis.set_major_locator(MaxNLocator(integer=True))
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].plot(years, stock_prices_microsoft)
axes[1].plot(years, stock_prices_apple)
```
